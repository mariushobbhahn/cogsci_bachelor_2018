```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=np.array([2,4,20])
sparse1_noise_mean = ([1, 0.97, 0.62])
sparse4_noise_mean = ([0.97, 0.94, 0.49])
noise_mean = np.array([0.97, 0.93, 0.43])
sparse_comparison_mean = np.array([0.98, 0.94, 0.61])
comparison_mean = np.array([1, 0.94, 0.97])

plt.plot(x, sparse1_noise_mean, label='sparse1_noise_mean', marker='^')
plt.plot(x, sparse4_noise_mean, label='sparse4_noise_mean', marker='^')
plt.plot(x, noise_mean, label='noise_mean', marker='o')
plt.plot(x, sparse_comparison_mean, label='sparse_comparison_mean', marker='x')
plt.plot(x, comparison_mean, label='comparison_mean', marker='x')
plt.xlabel('accuracy')
plt.ylabel('dimensions')
plt.legend()
plt.title('comparison of sparse and normal training')
#plt.plot(, label=sparse1_noise_mean)
```
